### Split Email and Phone Number

```python
clientParticular[clientParticular.FirstName == 'Zahraaaa']
```

```python
clientParticular['fullname'] = clientParticular['FirstName'] + clientParticular['LastName']
```

```python
clientParticular[clientParticular['fullname'].isnull()]
```

```python
clientParticular = clientParticular[clientParticular['FirstName'].notna()]
```

```python
clientParticular = clientParticular[clientParticular['PhoneNumber'].notna()]
```

```python
clientParticular['fullname'] = clientParticular['fullname'].apply(lambda x: x.lower())
```

```python
import string
import re
clientParticular['fullname'] = clientParticular['fullname'].apply(lambda x: re.sub(r'[^\w\s]', '', x))
```

```python
def splitContact(numemail):
    match = re.search('[a-zA-Z]', numemail)
    index = match.start()
#     print(index)
    part1 = numemail[:index]
    part2 = numemail[index:]
    return part1, part2
```

```python
for index, row in clientParticular.iterrows():
    num, email = splitContact(row.PhoneNumber)
    clientParticular.at[index,'PhoneNumber'] = num
    clientParticular.at[index,'EmailAddress'] = email
```

```python
indexToRem = []
for index, row in clientParticular.iterrows():
    if row.fullname not in row.EmailAddress:
        indexToRem.append(index)
```

```python
indexToRem
```

### Remove duplicate names

```python
clientParticular = clientParticular.drop(indexToRem)
```

### Impute High Value Data using Regression Imputation

```python
clientParticular = clientParticular.reset_index(drop=True)
```

```python
clientParticular[clientParticular.HighValueIndividual == "No"]
```

```python
clientParticular['HighValueIndividual'] = clientParticular['HighValueIndividual'].fillna(0)
```

```python
for index, row in clientParticular.iterrows():
    if row.HighValueIndividual == 0:
        if row.NetYearlyIncome < 300000:
            clientParticular.at[index, "HighValueIndividual"] = "No"
        else:
            clientParticular.at[index, "HighValueIndividual"] = "Yes"
```

### Impute Marriage Status

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
clientParticular.columns
```

```python
clientParticular_noage = clientParticular.drop(['Age'], axis = 1)
```

```python
def male(x):
    if x == "Male":
        return 1
    return 0
```

```python
clientParticular_noage['Male'] = clientParticular_noage['Gender'].apply(lambda x: male(x))
```

```python
def married(x):
    if x == "Married":
        return 1
    return 0
```

```python
clientParticular_noage['Married'] = clientParticular_noage['MarriageStatus'].apply(lambda x: married(x))
```

```python
# Split data into two groups: one with missing values and one without
missing_data = clientParticular_noage[clientParticular_noage['MarriageStatus'].isnull()]
missing_data
```

```python
non_missing_data = clientParticular_noage.dropna()
```

```python
non_missing_data
```

```python
reg_model = LinearRegression()
X = non_missing_data[['Male', 'NetYearlyIncome']]  # replace with your independent variables
y = non_missing_data['Married']
reg_model.fit(X, y)
```

```python
# Predict missing values
X_missing = missing_data[['Male', 'NetYearlyIncome']]  # replace with your independent variables
predicted_values = reg_model.predict(X_missing)
missing_data["Married"] = predicted_values
```

```python
def imputeStatus(x):
    if x < 0.5:
        return "Married"
    return "Single"
```

```python
missing_data['MarriageStatus'] = missing_data['Married'].apply(lambda x: imputeStatus(x))
```

```python
imputed_data = pd.concat([missing_data, non_missing_data])
```

```python
imputed_data
```

```python
rsimputed = pd.merge(imputed_data, clientParticular.drop(['MarriageStatus', 'PhoneNumber', 'NetYearlyIncome', 'EmailAddress', 'fullname', 'HighValueIndividual','Gender'], axis=1), on=['FirstName','LastName'],how='left')
```

```python
rsimputed
```

### Impute Age

```python
missing_data = rsimputed[rsimputed["Age"].isnull()]
non_missing_data = rsimputed.dropna()
```

```python
reg_model = LinearRegression()
X = non_missing_data[["NetYearlyIncome", "Male", "Married"]]  # replace with your independent variables
y = non_missing_data["Age"]
reg_model.fit(X, y)
```

```python
X_missing = missing_data[["NetYearlyIncome", "Male", "Married"]]  # replace with your independent variables
predicted_values = reg_model.predict(X_missing)
missing_data["Age"] = predicted_values
```

# Why Data Migration is necessary?

### Upgrading or replacing systems: When organizations upgrade or replace their software or hardware systems, they often need to migrate data from the old system to the new one. This ensures that important data is not lost during the transition and can continue to be used in the new system.

### Consolidating data: Organizations often have multiple databases or data sources that need to be combined into a single system. Data migration is necessary to ensure that all data is included in the new system and is accessible from a single location.

### Business mergers and acquisitions: When companies merge or are acquired, they may have different systems and data sources that need to be integrated. Data migration is necessary to ensure that all data is available to the new organization and can be used to make informed business decisions.

### Compliance and regulatory requirements: Organizations may need to migrate data to comply with legal and regulatory requirements, such as data retention laws, data protection regulations, and industry-specific standards.

### Performance improvements: Data migration can also be necessary to improve system performance. For example, data may need to be moved to a more powerful server or storage system to improve processing speed or reduce latency.

### Plan the migration: This step involves identifying the objectives of the migration, the scope of the migration, and the timeline for the migration. It is important to have a clear understanding of what is being migrated, the expected outcome, and any constraints that may affect the migration.

### Choose the target database: Identify the target database and determine if there are any limitations or compatibility issues. For example, if the database being migrated is an SQL Server database, the target database may need to be a SQL Server database as well.

### Create a migration plan: Create a detailed plan for the migration, which should include the following:

### Backup the database: Backup the database to ensure that there is a copy of the data in case something goes wrong during the migration.

### Test the migration: Test the migration on a test environment to ensure that the migration plan is accurate and that there are no unforeseen issues.

### Determine the migration method: Choose the best migration method for the database being migrated. This may include tools provided by the database vendor or third-party migration tools.

### Migrate the schema: Migrate the schema of the database to the target database. This includes creating tables, indexes, stored procedures, and triggers.

### Migrate the data: Migrate the data from the source database to the target database. This may involve extracting data from the source database, transforming the data if necessary, and loading it into the target database.

### Verify the migration: Verify that the migration was successful by checking the data in the target database against the source database.

### Communicate the migration: Communicate the migration plan and timeline to all stakeholders, including users, developers, and administrators.

### Perform the migration: Execute the migration plan, carefully following each step of the plan.

### Verify the migration: Verify that the migration was successful by checking the data in the target database against the source database.

### Test the new database: Test the new database to ensure that it is working properly and that all applications that rely on it are functioning as expected.

### Decommission the old database: Once the new database has been tested and is working properly, decommission the old database.

### Monitor the new database: Monitor the new database to ensure that it is performing as expected and that there are no issues that need to be addressed.
